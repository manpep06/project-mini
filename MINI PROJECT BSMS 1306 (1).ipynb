## Background of Study

Video gaming has become a popular activity among students due to the rapid advancement of digital technology. While gaming can provide entertainment and relaxation, excessive gaming may affect students' stress levels, study habits, sleep patterns, and academic performance.

Factors such as study time, sleep quality, and stress management play important roles in academic success. Students who spend too much time gaming may have less time for studying and resting, which can negatively impact their grades and overall well-being.

Therefore, this study aims to examine the relationship between gaming addiction and stress levels, investigate the influence of study hours and gaming hours on academic performance, and analyze the impact of sleep hours and gaming activity on students' grades through data analytics techniques.

## OBJECTIVE

1. To analyze the relationship between gaming addiction score and stress level among students.
2. To identify whether study hours and gaming hours influence academic performance.
3. To analyze the impact of sleep hours and gaming activity on students’ grades.

### Data analysis

#### Objective 1

```python
# Average addiction score for each stress level

addiction_stress = df.groupby('Stress Level')['Addiction Score'].mean()

print(addiction_stress)
```

#### Objective 2

```python
# study hour categories

df['Study_Category'] = pd.cut(
    df['Study Hours'],
    bins=[0,2,4,6,24],
    labels=['Low','Moderate','High','Very High']
)

# Average grades for each category

study_grades = df.groupby('Study_Category')['Grades'].mean()

print(study_grades)

# gaming hour categories

df['Gaming_Category'] = pd.cut(
    df['Gaming Hours'],
    bins=[0,2,4,6,24],
    labels=['Low','Moderate','High','Very High']
)

# Average grades for each category

gaming_grades = df.groupby('Gaming_Category')['Grades'].mean()

print(gaming_grades)


df.groupby('Study_Category')['Grades'].agg(
    ['count', 'mean', 'min', 'max', 'std']
)

df.groupby('Gaming_Category')['Grades'].agg(
    ['count', 'mean', 'min', 'max', 'std']
)
```

#### Objective 3

```python
# sleep categories

df['Sleep_Category'] = pd.cut(
    df['Sleep Hours'],
    bins=[0,6,8,24],
    labels=['Low Sleep','Adequate Sleep','High Sleep']
)

# Average grades by sleep category

sleep_grades = df.groupby('Sleep_Category')['Grades'].mean()

print(sleep_grades)
```

### Plotting Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("GamingData.csv")

# Count students in each stress level
stress_counts = df['stress_level'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(stress_counts,
        labels=stress_counts.index,
        autopct='%1.1f%%')

plt.title('Distribution of Stress Levels Among Students')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("GamingData.csv")

# Create study hour categories
df['study_category'] = pd.cut(
    df['study_hours'],
    bins=[0, 2, 4, 6, 24],
    labels=['Low', 'Moderate', 'High', 'Very High']
)

# Calculate average grades
avg_grades = df.groupby('study_category')['grades'].mean()

plt.figure(figsize=(8,6))
plt.bar(avg_grades.index, avg_grades.values)

plt.title('Average Grades by Study Hours Category')
plt.xlabel('Study Hours Category')
plt.ylabel('Average Grades')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("GamingData.csv")

# Categorize sleep hours
df['sleep_category'] = pd.cut(
    df['sleep_hours'],
    bins=[0, 6, 24],
    labels=['Low Sleep', 'High Sleep']
)

plt.figure(figsize=(8,6))

df[df['sleep_category'] == 'Low Sleep']['grades'].plot(
    kind='density',
    label='Low Sleep'
)

df[df['sleep_category'] == 'High Sleep']['grades'].plot(
    kind='density',
    label='High Sleep'
)

plt.title('Grade Distribution by Sleep Category')
plt.xlabel('Grades')
plt.ylabel('Density')

plt.legend()
plt.grid(True)

plt.show()
```

### Discussion

The analysis indicates that gaming behavior is related to students' academic performance and well-being. Students with higher gaming addiction scores generally tend to experience higher stress levels, suggesting a possible relationship between excessive gaming and stress.

The findings also show that study habits play an important role in academic achievement. Students who spend more time studying tend to achieve better grades, while excessive gaming may reduce the time available for academic activities.

In addition, sleep appears to be an important factor influencing academic performance. Students who obtain sufficient sleep generally perform better than those with inadequate sleep. Overall, the results highlight the importance of maintaining a balance between gaming, studying, and sleeping to support both academic success and personal well-being.

### Conclusion

This study analyzed the relationship between gaming behavior, stress levels, study habits, sleep patterns, and academic performance among students. The findings indicate that gaming addiction is associated with stress levels, while study hours, gaming activity, and sleep duration may influence academic achievement.

Students who maintain a healthy balance between gaming, studying, and sleeping tend to achieve better academic outcomes. Excessive gaming may contribute to increased stress and reduced time for studying and resting, which can negatively affect performance. On the other hand, sufficient study time and adequate sleep are important factors that support academic success.

In conclusion, responsible gaming habits and effective time management are essential for maintaining both academic performance and personal well-being. The results of this study provide useful insights for students, educators, and parents in promoting a balanced and healthy lifestyle.
